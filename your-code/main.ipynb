How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
np.sort(fitbit["Months"].unique())
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
steps_month = fitbit.groupby(["Months_encoded"])["Steps"].median()
month_index = steps_month.index


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
steps_ww = fitbit.groupby(["Work_or_Weekend"])["Steps"].median()
steps_index = steps_ww.index

fig3, ax3= plt.subplots(nrows=1, ncols=2, figsize=(12,7))
ax3[0].bar(month_index, steps_month, color="mediumturquoise", label="Median steps by month")
plt.sca(ax3[0])
plt.xticks(ticks=range(1,13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct',
       'Nov', 'Dec'], rotation=20)
ax3[0].set_title("Median steps by month", fontsize=15)


ax3[1].bar(steps_index, steps_ww, label="Median steps by weekdays or weekends")
plt.sca(ax3[1])
plt.xticks(ticks=range(2), labels=["Weekends", "Workdays"])
ax3[1].set_title("Median steps by weekdays or weekends", fontsize=15)
```
